# GravBalancer harness — visual run (v10.9)

Thin wrapper over `harness.main()`. The canonical artifact is the
self-contained `harness_report.html` (overwritten each run; history = git).
This notebook just runs the same thing and shows the panels inline.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
%matplotlib inline
import matplotlib.pyplot as plt
from harness import run_all, print_summary, _scenario_panel, MODES, render_html_report
```

```python
all_results = {}
for mode_name, freeze in MODES.items():
    results = run_all(verbose=False, debug_freeze=freeze)
    print_summary(results, mode_name)
    all_results[mode_name] = results
```

```python
# Inline panels for every scenario x mode
for mode_name, results in all_results.items():
    for name, r in results.items():
        fig = _scenario_panel(r['history'], name, mode_name)
        plt.show()
```

```python
# Optionally also (re)write the canonical HTML report
render_html_report(all_results, out_path='../harness_report.html')
```
